### Store Part I results into DataFrame

```python
# Reference the file where the CSV is located
weather_path = "../instructions/output_data/cities.csv"

# Import the data into a Pandas DataFrame
weather_data = pd.read_csv(weather_path)
weather_data.head()
```

```python
# Remove Unnamed column
weather_data = weather_data[["City","Country","Cloudiness","Humidity","Lat","Lng","Max Temp","Wind Speed"]]
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# configure gmaps 
gmaps.configure(api_key=g_key)
```

```python
# set coordinates
coordinates = weather_data[["Lat", "Lng"]].astype(float)

# convert the new DataFrame to a list that can be passed 
coordinates = coordinates.values.tolist()

Humidity_heatmap = weather_data["Humidity"].astype(float)

# Create a humidity Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(coordinates, weights=Humidity_heatmap, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```
